```python
# Part 1: Data Governance Policies
# Data governance ensures that data is properly managed, organized, and controlled in an organization.

# Import Library
import pandas as pd

# load dataset
df = pd.read_csv("student.csv")

# view dataset
print(df.head())

# check structure
print(df.info())

# check missing values
print(df.isnull().sum())
```

```python
# Part 2: Data Privacy Policies
# Data privacy ensures that personal information is protected.

# 6.check personal columns
print(df.columns)

# 7.example: remove personal identification column
df_privacy = df.drop(columns=["Student_ID"])

# 8.dispaly Dataset
print(df_privacy.head())
```

```python
# Part 3: Data Security Standards
# Data security protects data from unauthorized access or cyber attacks.

# 9.save secured copy
df.to_csv("secured_student_data.csv", index=False)

# 10: Display confirmation message
print("Secure file successfully")
```

```python
# Part 4: Ethical Data Usage
# Data should be used fairly without discrimination or misuse.

# 11.average score by gender
gender_score = df.groupby("Gender")["Final_Score"].mean()

# 12.Dispaly gender_score
print(gender_score)
```

```python
# Part 5: Data Quality Standards
# Data quality ensures data is accurate, consistent, and reliable.

# 13.check missing values
print(df.isnull().sum())

# 14.remove duplicates
df = df.drop_duplicates()

# 15.describe dataset
print(df.describe())

# 16.check data types
print(df.dtypes)
```
